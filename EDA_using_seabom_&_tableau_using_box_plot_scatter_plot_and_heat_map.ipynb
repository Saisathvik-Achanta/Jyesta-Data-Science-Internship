# Task
Perform exploratory data analysis (EDA) on a random dataset using seaborn and Tableau, demonstrating the analysis with box plots, scatter plots, and heatmaps.

## Select and load data

### Subtask:
Choose a suitable random dataset and load it into a pandas DataFrame.


**Reasoning**:
Import the pandas library and load a suitable random dataset into a pandas DataFrame.



```python
import pandas as pd
import numpy as np

# Generate a random dataset
data = {
    'A': np.random.rand(100),
    'B': np.random.randn(100),
    'C': np.random.randint(0, 100, 100),
    'D': np.random.choice(['X', 'Y', 'Z'], 100)
}
df = pd.DataFrame(data)

# Display the first few rows of the DataFrame
print(df.head())
```

## Initial data exploration

### Subtask:
Perform basic data exploration to understand the structure, types, and summary statistics of the dataset.


**Reasoning**:
Perform basic data exploration using `.info()`, `.describe()`, `.head()`, and `.tail()` methods to understand the structure, types, and summary statistics of the dataset as per the instructions.



```python
# Get a concise summary of the DataFrame
df.info()

# Generate descriptive statistics
display(df.describe())

# Display the first 5 rows
display(df.head())

# Display the last 5 rows
display(df.tail())
```

## Generate visualizations with seaborn

### Subtask:
Create a box plot, scatter plot, and heatmap using seaborn to visualize the data.


**Reasoning**:
Create the required visualizations (box plot, scatter plot, and heatmap) using seaborn to visualize the data as requested in the instructions.



```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a box plot of column 'C' grouped by column 'D'
plt.figure(figsize=(8, 6))
sns.boxplot(x='D', y='C', data=df)
plt.xlabel('Category')
plt.ylabel('Value of C')
plt.title('Box Plot of C by Category D')
plt.show()

# Create a scatter plot of columns 'A' and 'B'
plt.figure(figsize=(8, 6))
sns.scatterplot(x='A', y='B', data=df)
plt.xlabel('Value of A')
plt.ylabel('Value of B')
plt.title('Scatter Plot of A vs B')
plt.show()

# Calculate the correlation matrix of the numerical columns
numerical_cols = df[['A', 'B', 'C']]
correlation_matrix = numerical_cols.corr()

# Create a heatmap of the calculated correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Columns')
plt.show()
```
